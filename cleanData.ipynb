```python
import pandas as pd
import numpy as np
from dateutil.parser import parse
from datetime import datetime
```

```python
E1=pd.read_csv("employee_data.csv")
```

```python
null_columns = E1.isnull().sum()
print("Columns with NULL values:")
print(null_columns[null_columns > 0])
```

```python
# Fill the ExitDate with current data
E1['ExitDate'] = E1['ExitDate'].apply(lambda x: parse(x, dayfirst=True) if pd.notnull(x) else None)
current_date = datetime.now().strftime('%Y-%m-%d') 
E1['ExitDate'].fillna(current_date,inplace=True)


#fill null with still working
E1['TerminationDescription'].fillna("Still Working" ,inplace=True)

#change the Unk to Stillworking
E1.loc[E1['TerminationType'] == 'Unk', 'TerminationType'] = 'Still Working'
```

```python
E1['StartDate'] = E1['StartDate'].apply(lambda x: parse(x, dayfirst=True) if pd.notnull(x) else None)
```

```python
E1['Age'] = E1['DOB'].astype(str)
E1['Age'] = E1['Age'].apply(lambda x: parse(x, dayfirst=True) if pd.notnull(x) else None)

E1['Age']=E1['Age'].dt.year
E1['Age']=E1['Age'].astype(int)
current_year = pd.Timestamp('today').year
E1['Age'] = current_year - E1['Age']

E1 = E1.drop('DOB', axis=1)
```

```python
business_unit_mapping = {
    'BPC': 'Business Planning and Control',
    'CCDR': 'Corporate Communications and Data Relations',
    'EW': 'Employee Wellness',
    'MSC': 'Management Services Center',
    'NEL': 'New Employee Learning',
    'PL': 'Product Line',
    'PYZ': 'Product Yield and Zoning',
    'SVG': 'Strategic Ventures Group',
    'TNS': 'Technology and Network Services',
    'WBL': 'Work-Life Balance'
}


# Replace values in the 'BusinessUnit' column using the mapping
E1['BusinessUnit'] = E1['BusinessUnit'].map(business_unit_mapping).fillna(E1['BusinessUnit'])
```

```python
state_mapping = {
    'AL': 'Alabama',
    'AZ': 'Arizona',
    'CA': 'California',
    'CO': 'Colorado',
    'CT': 'Connecticut',
    'FL': 'Florida',
    'GA': 'Georgia',
    'ID': 'Idaho',
    'IN': 'Indiana',
    'KY': 'Kentucky',
    'MA': 'Massachusetts',
    'ME': 'Maine',
    'MT': 'Montana',
    'NC': 'North Carolina',
    'ND': 'North Dakota',
    'NH': 'New Hampshire',
    'NV': 'Nevada',
    'NY': 'New York',
    'OH': 'Ohio',
    'OR': 'Oregon',
    'PA': 'Pennsylvania',
    'RI': 'Rhode Island',
    'TN': 'Tennessee',
    'TX': 'Texas',
    'UT': 'Utah',
    'VA': 'Virginia',
    'VT': 'Vermont',
    'WA': 'Washington'
}

# Fetch data from the E1 table


# Replace values in the 'State' column using the state mapping
E1['State'] = E1['State'].map(state_mapping).fillna(E1['BusinessUnit'])
```

```python
payzone_mapping = {
    'Zone A': 'Entry-Level Salary',
    'Zone B': 'Mid-Level Salary',
    'Zone C': 'Senior-Level Salary'
}

# Replace values in the 'PayZone' column using the payzone mapping
E1['PayZone'] = E1['PayZone'].map(payzone_mapping).fillna(E1['PayZone'])
```
